# Shi-Tomasi Corner Detector & Good Features to Track

OpenCV has a function, cv2.goodFeaturesToTrack(). It finds N strongest corners in the image by Shi-Tomasi method

In below example, we will try to find 25 best corners

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('images.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(gray,25,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,255,-1)

plt.imshow(img),plt.show()
```

This function is more appropriate for tracking.

###  In below example, we will try to find 5 best corners

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('images.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(gray,5,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,255,-1)

plt.imshow(img),plt.show()
```
